```python


import nltk
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords  # Add this import statement
from nltk.stem import WordNetLemmatizer
import string
import contractions
nltk.download('punkt')
nltk.download('stopwords')
nltk.download('wordnet')
# Load SQuAD dataset
# squad_data = load_dataset("squad")

# Function to perform lexical processing with lemmatization and punctuation removal

def expand_contractions(text):
    # Expand contractions using the contractions library
    expanded_text = contractions.fix(text)
    return expanded_text

def lexical_processing_with_lemmatization_and_punctuation_removal(example):
    question = example['question']
    context = example['context']
    answer = example['answers']['text'][0]  # Assuming you want to use the first answer

    question = expand_contractions(question)
    context = expand_contractions(context)
    answer = expand_contractions(answer)

    # Tokenize, lowercase, and remove punctuation
    question_tokens = [token.lower() for token in word_tokenize(question) if token.isalnum()]
    context_tokens = [token.lower() for token in word_tokenize(context) if token.isalnum()]
    answer_tokens = [token.lower() for token in word_tokenize(answer) if token.isalnum()]

    # Remove stop words
    stop_words = set(stopwords.words('english'))
    question_tokens = [token for token in question_tokens if token not in stop_words]
    context_tokens = [token for token in context_tokens if token not in stop_words]
    answer_tokens = [token for token in answer_tokens if token not in stop_words]

    # Lemmatization
    lemmatizer = WordNetLemmatizer()
    question_tokens = [lemmatizer.lemmatize(token) for token in question_tokens]
    context_tokens = [lemmatizer.lemmatize(token) for token in context_tokens]
    answer_tokens = [lemmatizer.lemmatize(token) for token in answer_tokens]

    return {
        'question_tokens': question_tokens,
        'context_tokens': context_tokens,
        'answer_tokens': answer_tokens,
    }

# Apply lexical processing with lemmatization and punctuation removal to the entire SQuAD dataset
processed_data_with_lemmatization_and_punctuation_removal = [
    lexical_processing_with_lemmatization_and_punctuation_removal(example) for example in train_dataset
]

# Create a DataFrame for better visualization
df = pd.DataFrame(processed_data_with_lemmatization_and_punctuation_removal)

# Save the processed data to a CSV file
df.to_csv('squad_lexical_processed.csv', index=False)
```

```python
# Install necessary libraries
!pip install --quiet flashtext==2.7
!pip install git+https://github.com/boudinfl/pke.git
!pip install nltk

# Import the libraries
import pke
import nltk
nltk.download('stopwords')

from nltk.corpus import stopwords
import string
import traceback

def get_nouns_multipartite(content):
    out=[]
    try:
        extractor = pke.unsupervised.MultipartiteRank()
        extractor.load_document(input=content,language='en')
        #    not contain punctuation marks or stopwords as candidates.
        pos = {'PROPN','NOUN', 'ADJ'}
        #pos = {'PROPN','NOUN'}
        stoplist = list(string.punctuation)
        stoplist += ['-lrb-', '-rrb-', '-lcb-', '-rcb-', '-lsb-', '-rsb-']
        stoplist += stopwords.words('english')

        extractor.candidate_selection(pos=pos , stoplist=stoplist)
        # 4. build the Multipartite graph and rank candidates using random walk,
        #    alpha controls the weight adjustment mechanism, see TopicRank for
        #    threshold/method parameters.
        extractor.candidate_weighting(alpha=1.1,
                                      threshold=0.75,
                                      method='average')
        keyphrases = extractor.get_n_best(n=15)

        for val in keyphrases:
            out.append(val[0])
    except:
        out = []
        traceback.print_exc()

    return out
```

```python

def get_keywords(originaltext):
  keywords = get_nouns_multipartite(originaltext)
  return keywords

extracted_keywords = get_keywords(text)
extracted_keywords
```
